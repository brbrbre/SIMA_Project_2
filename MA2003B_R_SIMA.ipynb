Ahora se hará un análisis exploratorio de los datos con la limpieza ya realizada. Como primer putno se toma en cuenta las variables 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import copy

# Lista de banderas inválidas (Tabla 3)
invalid_flags = ['p', 'c', 'd', 'b', 'm', 'l', 'z', 'o', 's', 'r', 'e', 'a', 'f', 'h', 'n', 'x']
all_quant_vars = ['CO', 'NO', 'NO2', 'NOX', 'O3', 'PM10', 'PM2.5', 'SO2', 'PRS', 'RAINF', 'RH', 'SR', 'TOUT', 'WSR', 'WDR']
key_vars = ['NO2', 'PM2.5', 'SO2']  # Variables clave para tráfico/refinería

# Función para analizar nulos y calidad
def analyze_dataset(dataset, dataset_name):
    print("="*80, f"{dataset_name}")
    null_summary = {}
    total_rows = 0
    valid_stations = []
    
    # Copia profunda para no modificar el original
    dataset_copy = copy.deepcopy(dataset)
    
    for station in dataset_copy:
        if station != 'CATÁLOGO':  # Ignorar CATÁLOGO
            df = dataset_copy[station].copy()
            
            # 1. Diagnóstico inicial: Verificar datos crudos
            print("-"*50, station)
            print("-"*25, "Datos Crudos (Primeras 5 filas)")
            print(df.head())
            print("-"*25, "Tipos de Datos Iniciales")
            print(df.info())
            print("-"*25, "Nulos Iniciales")
            null_counts_initial = df.isnull().sum()
            print(pd.DataFrame({'Nulos': null_counts_initial}))
            
            # Verificar valores únicos en columnas cuantitativas
            quant_vars = [col for col in all_quant_vars if col in df.columns]
            if not quant_vars:
                print(f"Advertencia: {station} no tiene variables cuantitativas. Omitiendo.")
                continue
            print("-"*25, "Valores Únicos")
            for col in quant_vars:
                print(f"{col}: {df[col].unique()[:10]}")  # Primeros 10 valores
            
            # Convertir a numérico
            for col in quant_vars:
                df[col] = pd.to_numeric(df[col], errors='coerce')
            
            # Contar nulos después de conversión
            print("-"*25, "Nulos después de Conversión")
            null_counts = df[quant_vars].isnull().sum()
            null_percent = (null_counts / df.shape[0] * 100).round(2)
            print(pd.DataFrame({'Nulos': null_counts, 'Porcentaje (%)': null_percent}))
            
            # Verificar banderas si existen
            if 'bandera' in df.columns:
                print("ENTRE A ESTE CICLO##################")
                print("-"*25, "Banderas")
                print(df['bandera'].value_counts())
                invalid_count = df['bandera'].isin(invalid_flags).sum()
                print(f"Datos con banderas inválidas: {invalid_count} ({invalid_count/df.shape[0]*100:.2f}%)")
                df = df[~df['bandera'].isin(invalid_flags)]
            
            # Aplicar reglas de Tabla 3
            df = df[df[quant_vars].ge(0).all(axis=1)]  # Negativos
            if 'PM10' in quant_vars and 'PM2.5' in quant_vars:
                print("ENTRE A ESTE CICLO################## PM10 PM2.5")
                df = df[df['PM2.5'] <= df['PM10']]  # PM2.5 ≤ PM10
                df = df[(df['PM10'] >= 5) & (df['PM2.5'] >= 5)]  # PM < 5
            if 'PM10' in quant_vars:
                print("ENTRE A ESTE CICLO################## PM10")
                df = df[df['PM10'] <= 900]  # PM10 > 900
            if 'CO' in quant_vars:
                df = df[df['CO'] >= 0.05]  # CO < 0.05
                print("ENTRE A ESTE CICLO################## CO")
            if 'TOUT' in quant_vars and 'PRS' in quant_vars:
                print("ENTRE A ESTE CICLO################## TOUT")
                df['temp_diff'] = df['TOUT'].diff().abs()
                df['pres_diff'] = df['PRS'].diff().abs()
                df = df[(df['temp_diff'].isna() | (df['temp_diff'] <= 10)) & 
                        (df['pres_diff'].isna() | (df['pres_diff'] <= 10))]
                df = df.drop(columns=['temp_diff', 'pres_diff'], errors='ignore')
            
            # Guardar DataFrame limpio en copia
            dataset_copy[station] = df
            total_rows += df.shape[0]
            null_summary[station] = {'counts': null_counts, 'percent': null_percent}
            valid_stations.append(station)
    
    # Resumen global
    print("-"*50, "Resumen Global")
    print(f"Total filas (tras limpieza): {total_rows}")
    total_nulls = sum([null_summary[station]['counts'].sum() for station in null_summary])
    print(f"Total nulos en variables cuantitativas: {total_nulls}")
    print(f"Porcentaje nulos global: {(total_nulls/(total_rows*len(quant_vars))*100):.2f}%")
    
    return null_summary, valid_stations, dataset_copy

# Analizar ambos datasets
null_summary_2023_2024, valid_stations_2023_2024, db_2023_2024_clean = analyze_dataset(db_2023_2024, "Dataset 2023-2024")
null_summary_2020_2022, valid_stations_2020_2022, db_2020_2022_clean = analyze_dataset(db_2020_2022, "Dataset 2020-2022")

# Función para generar el top de columnas con más nulos
def top_nulls(null_summary, dataset_name, top_n=5):
    print("="*80, f"Top Columnas con Más Nulos en {dataset_name}")
    # Aggregar nulos por columna a través de todas las estaciones
    aggregated_nulls = pd.DataFrame(null_summary).sum(axis=1)  # Sum nulos por columna
    aggregated_percent = pd.DataFrame({station: null_summary[station]['percent'] for station in null_summary}).mean(axis=1)  # Promedio % por columna
    
    # DataFrame con nulos y porcentajes agregados
    top_df = pd.DataFrame({'Nulos Totales': aggregated_nulls, 'Porcentaje Promedio (%)': aggregated_percent.round(2)})
    top_df = top_df.sort_values('Nulos Totales', ascending=False).head(top_n)  # Top N con más nulos
    
    print(top_df)
    top_df.to_csv(f'top_nulls_{dataset_name.replace(" ", "_")}.csv', index=True)
    
    return top_df

# Generar top para ambos datasets
top_2023_2024 = top_nulls(db_2023_2024_variablesImportantes, "Dataset 2023-2024")
top_2020_2022 = top_nulls(db_2020_2022_variablesImportantes, "Dataset 2020-2022")

# Comparar datasets
print("="*80, "Comparación de Datasets")
print("Porcentaje de nulos en variables clave (NO2, PM2.5, SO2):")
for station in set(valid_stations_2023_2024) & set(valid_stations_2020_2022):
    key_nulls_2023_2024 = null_summary_2023_2024[station]['percent'][key_vars]
    key_nulls_2020_2022 = null_summary_2020_2022[station]['percent'][key_vars]
    print(f"\nEstación: {station}")
    print("2023-2024:", key_nulls_2023_2024.to_dict())
    print("2020-2022:", key_nulls_2020_2022.to_dict())

# Decisión basada en nulos en key_vars
avg_nulls_2023_2024 = np.nanmean([null_summary_2023_2024[station]['percent'][key_vars].mean() 
                                  for station in valid_stations_2023_2024 
                                  if station in null_summary_2023_2024 and key_vars[0] in null_summary_2023_2024[station]['percent']])
avg_nulls_2020_2022 = np.nanmean([null_summary_2020_2022[station]['percent'][key_vars].mean() 
                                  for station in valid_stations_2020_2022 
                                  if station in null_summary_2020_2022 and key_vars[0] in null_summary_2020_2022[station]['percent']])
print("\nDecisión:")
if not np.isnan(avg_nulls_2023_2024) and not np.isnan(avg_nulls_2020_2022):
    if avg_nulls_2023_2024 < avg_nulls_2020_2022:
        print(f"Recomiendo usar db_2023_2024 para modelado (menos nulos: {avg_nulls_2023_2024:.2f}% vs. {avg_nulls_2020_2022:.2f}%)")
        print("Usa db_2020_2022 para EDA comparativo (COVID-19, políticas).")
    else:
        print(f"Recomiendo usar db_2020_2022 para modelado (menos nulos: {avg_nulls_2020_2022:.2f}% vs. {avg_nulls_2023_2024:.2f}%)")
        print("Usa db_2023_2024 para EDA comparativo.")
else:
    print("No se pudo calcular nulos promedio debido a datos faltantes. Revisa las estaciones.")

# Gráfico de frecuencias de promedios diarios
dfs = []
for station in db_2023_2024_clean:
    if station in valid_stations_2023_2024:
        df = db_2023_2024_clean[station].copy()
        df['estacion'] = station
        dfs.append(df)
for station in db_2020_2022_clean:
    if station in valid_stations_2020_2022:
        df = db_2020_2022_clean[station].copy()
        df['estacion'] = station
        dfs.append(df)
df_all = pd.concat(dfs, ignore_index=True)
df_all['date'] = pd.to_datetime(df_all['date'], format='%d/%m/%y %H:%M', errors='coerce')

# Promedios diarios por estación
df_all['date_day'] = df_all['date'].dt.date
daily_means = df_all.groupby(['date_day', 'estacion'])[key_vars].mean().reset_index()

# Histogramas y boxplots
for var in key_vars:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=daily_means, x=var, hue='estacion', bins=30, kde=True, multiple='stack')
    plt.title(f'Histograma de Promedios Diarios de {var} por Estación')
    plt.savefig(f'hist_daily_station_{var}.png')
    plt.close()

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=daily_means, y=var, x='estacion')
    plt.title(f'Boxplot de Promedios Diarios de {var} por Estación')
    plt.xticks(rotation=90)
    plt.savefig(f'box_daily_station_{var}.png')
    plt.close()

# Guardar dataset limpio
df_all.to_csv('dataset_descriptivo.csv', index=False)
```
